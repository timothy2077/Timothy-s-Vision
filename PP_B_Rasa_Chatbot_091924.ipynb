### F. Convert CSV to nlu.yml

```python
import pandas as pd
import yaml
import os

# Load the CSV file
csv_file_path = "BP_MHS_V1.csv"

# Check if the CSV file exists
if not os.path.exists(csv_file_path):
    print(f"Error: The file '{csv_file_path}' does not exist in the current directory.")
else:
    df = pd.read_csv(csv_file_path)

    # Initialize NLU data structure
    nlu_data = {"version": "3.1", "nlu": []}

    # Convert CSV rows to NLU examples
    for i, row in df.iterrows():
        if "questions" in df.columns and "tag" in df.columns:
            intent = row["tag"]
            text = row["questions"]
            if intent and text:
                # Check if the intent already exists in nlu_data
                existing_intent = next((item for item in nlu_data["nlu"] if item["intent"] == intent), None)
                if existing_intent:
                    existing_intent["examples"] += f"\n- {text}"
                else:
                    nlu_data["nlu"].append({
                        "intent": intent,
                        "examples": f"- {text}"
                    })
        else:
            print(f"Error: Missing required columns in row {i}. Required columns: 'questions', 'tag'")

    # Add sample examples for 'greet' and 'goodbye' intents
    greet_examples = """
- hello
- hi
- hey
- good morning
- good evening
- hey there
"""
    goodbye_examples = """
- bye
- goodbye
- see you later
- have a nice day
- catch you later
- farewell
"""
    # Append 'greet' intent
    nlu_data["nlu"].append({
        "intent": "greet",
        "examples": greet_examples.strip()
    })

    # Append 'goodbye' intent
    nlu_data["nlu"].append({
        "intent": "goodbye",
        "examples": goodbye_examples.strip()
    })

    # Save the NLU data to nlu.yml inside the data directory
    if not os.path.exists("data"):
        os.makedirs("data")

    with open("data/nlu.yml", "w") as outfile:
        yaml.dump(nlu_data, outfile, sort_keys=False)

    print("Rasa NLU YAML format conversion completed and saved to `data/nlu.yml`.")
```

## 6. 모델을 통해 예측하기_Option(predict_intent)
